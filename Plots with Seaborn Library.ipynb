```python
df['custAge']=df['custAge'].fillna(df['custAge'].mean())
```

### Scatter Plot

```python
sns.relplot(x="custAge",y="nr.employed",data=df)
```

### Hue Plot

```python
sns.relplot(x="custAge",y="nr.employed",hue="responded",data=df)
```

```python
sns.relplot(x="custAge",y="nr.employed",hue="day_of_week",data=df)
```

```python
sns.relplot(x="custAge",y="nr.employed",size="responded",data=df)
```

### Categorical Plot

```python
sns.catplot(x="responded",y="nr.employed",data=df)
```

```python
sns.catplot(x="responded",y="nr.employed",jitter=False,data=df)
```

##### Categorical Plot with Hue

```python
sns.catplot(x="responded",y="nr.employed",hue="loan",data=df)
```

##### Categorical swarm Plot

```python
sns.catplot(x="responded",y="nr.employed",kind="swarm",data=df)
```

```python
sns.catplot(x="responded",y="nr.employed",hue="loan",kind="swarm",data=df)
```

### Boxplot

```python
sns.catplot(x="responded",y="nr.employed",kind="box",data=df)
```

##### Boxplot with Hue

```python
sns.catplot(x="responded",y="nr.employed",kind="box",hue="loan",data=df)
```

### Violin Plot

```python
sns.catplot(x="responded",y="nr.employed",kind="violin",hue="loan",data=df)
```

##### Violin Plot with Hue

### Barplot

### Pointplot

```python
sns.catplot(x="responded",y="nr.employed",kind="point",hue="loan",data=df)
```

```python
sns.distplot(df.custAge,kde=False)
```
